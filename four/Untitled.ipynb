```python
import pandas as pd

df=pd.read_csv(r'C:\Users\hp zbook\homewok\four\command.csv') 


print(df.head())
```

```python
df['spam']=df['Title'].apply(lambda x:1 if x=='spam'else 0)
df
```

```python
dataset=df.drop(['Title'],axis='columns',errors='ingnore')
dataset
```

```python
print(df.groupby('spam').describe())
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(dataset,test_size=0.2,random_state=42)
print(train.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
train_vector=vectorizer.fit_transform(train.Positive.fillna(''))
print(train_vector.toarray())
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(train_vector,train.spam)
```

```python
test_vector=vectorizer.transform(test.Positive.fillna(''))
print(test_vector.toarray())
```

```python
emails=['  Wonderful place to stay.']
emails_vector=vectorizer.transform(emails)
emails_vector
```

```python
predictions=model.predict(emails_vector)
print(predictions)
```
